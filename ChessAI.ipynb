```
import numpy as np

##CHESS FRAME

class frameboard():


  #Creates a gameboard in its original state
  def createboard():
    pieces = {'Rb1'   :   ['Rb1'  , 'Active', False], 
              'KNb1'  :   ['KNb1' , 'Active' ],
              'Bb1'   :   ['Bb1'  , 'Active'], 
              'Qb'    :   ['Qb'   , 'Active'], 
              'Ib'    :   ['Ib'   , 'Active', False], 
              'Bb2'   :   ['Bb2'  , 'Active'], 
              'KNb2'  :   ['Rb1'  , 'Active'], 
              'Rb2'   :   ['Rb2'  , 'Active', False],
              'Pb1'   :   ['Pb1'  , 'Active'],
              'Pb2'   :   ['Pb2'  , 'Active'],
              'Pb3'   :   ['Pb3'  , 'Active'],
              'Pb4'   :   ['Pb4'  , 'Active'],
              'Pb5'   :   ['Pb5'  , 'Active'],
              'Pb6'   :   ['Pb6'  , 'Active'],
              'Pb7'   :   ['Pb7'  , 'Active'],
              'Pb8'   :   ['Pb8'  , 'Active'],

              'Rw1'   :   ['Rw1'  , 'Active', False],
              'KNw1'  :   ['KNw1' , 'Active'],
              'Bw1'   :   ['Bw1'  , 'Active'], 
              'Qw'    :   ['Qw'   , 'Active'], 
              'Iw'    :   ['Iw'   , 'Active', False], 
              'Bw2'   :   ['Bw2'  , 'Active'], 
              'KNw2'  :   ['Rw1'  , 'Active'], 
              'Rw2'   :   ['Rw2'  , 'Active', False],
              'Pw1'   :   ['Pw1'  , 'Active'],
              'Pw2'   :   ['Pw2'  , 'Active'],
              'Pw3'   :   ['Pw3'  , 'Active'],
              'Pw4'   :   ['Pw4'  , 'Active'],
              'Pw5'   :   ['Pw5'  , 'Active'],
              'Pw6'   :   ['Pw6'  , 'Active'],
              'Pw7'   :   ['Pw7'  , 'Active'],
              'Pw8'   :   ['Pw8'  , 'Active'],

              'blk'   :   ['  '   , 'Active'] }                                            

    board = [['Rb1', 'KNb1', 'Bb1', 'Qb' , 'Ib' , 'Bb2', 'KNb2', 'Rb2' ],
             ['Pb1', 'Pb2' , 'Pb3', 'Pb4', 'Pb5', 'Pb6', 'Pb7' , 'Pb8' ],
             ['blk', 'blk' , 'blk', 'blk', 'blk', 'blk', 'blk' , 'blk' ],
             ['blk', 'blk' , 'blk', 'blk', 'blk', 'blk', 'blk' , 'blk' ],
             ['blk', 'blk' , 'blk', 'blk', 'blk', 'blk', 'blk' , 'blk' ],
             ['blk', 'blk' , 'blk', 'blk', 'blk', 'blk', 'blk' , 'blk' ],
             ['Pw1', 'Pw2' , 'Pw3', 'Pw4', 'Pw5', 'Pw6', 'Pw7' , 'Pw8' ],
             ['Rw1', 'KNw1', 'Bw1', 'Qw' , 'Iw' , 'Bw2', 'KNw2', 'Rw2' ]]

    board = np.asarray(board) #Formats into numpy array for easy display
    return board

class moves():


  #Checks the value of a location (can it be moved into, is there another piece there, if so, from team of piece or opposing team?)
  def check(move, piece, board):

    if move[0]<0 or move[0]>7 or move[1]<0 or move[1]>7:
      return 'impossible'

    spot = board[move[0], move[1]]
    piece = board[piece[0], piece[1]]

    if spot == 'blk':
      return 'blank'

    elif (('w' in piece) and ('b' in spot)) or (('b' in piece) and ('w' in spot)):
      return spot

    else:
      return 'impossible'
  def removeimp(possiblemoves):
    temp = []
    for move in possiblemoves:
      if move[2] != "impossible":
        temp.append(move) 
    return temp
  def pawn(piece, board):
    row = piece[0]
    column = piece[1]
    possiblemoves = []

    if 'b' in board[row, column]:
      if row == 1:
        rawpossiblemoves = [[row+1, column], [row+2, column]]
        killingmoves =  [[row+1, column+1], [row+1, column-1]] 
      else:
        rawpossiblemoves = [[row+1, column]]
        killingmoves = [[row+1, column+1], [row+1, column-1]]
      

    elif 'w' in board[row, column]:
      if row == 6:
        rawpossiblemoves = [[row-1, column], [row-2, column]]
        killingmoves = [[row-1, column+1], [row-1, column-1]]
      else:
        rawpossiblemoves = [[row-1, column]]
        killingmoves = [[row-1, column+1], [row-1, column-1]]
    for i in range(len(rawpossiblemoves)):
        if board[rawpossiblemoves[i][0], rawpossiblemoves[i][1]] == "blk":
          possiblemoves.append(rawpossiblemoves[i])
        if len(rawpossiblemoves) == 2 and board[rawpossiblemoves[0][0], rawpossiblemoves[0][1]]!= "blk":
          possiblemoves = []
    for i in range(len(killingmoves)):
      if board[killingmoves[i][0], killingmoves[i][1]] != "blk":
        possiblemoves.append(killingmoves[i])

    for possiblemove in possiblemoves:
      possiblemove.append(moves.check(possiblemove, piece, board))
    possiblemoves = moves.removeimp(possiblemoves)
    return possiblemoves
  
  def rook(piece, board):
    row = piece[0]
    column = piece[1]
    possiblemoves = []    
    for i in range(8):
      possiblemoves.append([row+i, column])
    for i in range(8):
      possiblemoves.append([row-i, column])
    for i in range(8):
      possiblemoves.append([row, column+i])
    for i in range(8):
      possiblemoves.append([row, column-i])

    for possiblemove in possiblemoves:
      possiblemove.append(moves.check(possiblemove, piece, board))
    
    for possiblemove in possiblemoves:
      
      if possiblemove[0]>row and possiblemove[1]==column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]>possiblemove[0] and possible[1]==possiblemove[1]:
            possible[2] = 'impossible'

      if possiblemove[0]<row and possiblemove[1]==column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]<possiblemove[0] and possible[1]==possiblemove[1]:
            possible[2] = 'impossible'    
      
      if possiblemove[0]==row and possiblemove[1]>column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]==possiblemove[0] and possible[1]>possiblemove[1]:
            possible[2] = 'impossible'

      if possiblemove[0]==row and possiblemove[1]<column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]==possiblemove[0] and possible[1]<possiblemove[1]:
            possible[2] = 'impossible'        
            
    possiblemoves = moves.removeimp(possiblemoves)
    return possiblemoves

  def bishop(piece, board):
    row = piece[0]
    column = piece[1]    
    possiblemoves = []
    for i in range(8):
      possiblemoves.append([row+i, column+i])
    for i in range(8):
      possiblemoves.append([row-i, column+i])
    for i in range(8):
      possiblemoves.append([row+i, column-i])
    for i in range(8):
      possiblemoves.append([row-i, column-i])


    for possiblemove in possiblemoves:
      possiblemove.append(moves.check(possiblemove, piece, board))

    for possiblemove in possiblemoves:
      
      if possiblemove[0]>row and possiblemove[1]>column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]>possiblemove[0] and possible[1]>possiblemove[1]:
            possible[2] = 'impossible'

      if possiblemove[0]<row and possiblemove[1]<column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]<possiblemove[0] and possible[1]<possiblemove[1]:
            possible[2] = 'impossible'    
      
      if possiblemove[0]<row and possiblemove[1]>column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]<possiblemove[0] and possible[1]>possiblemove[1]:
            possible[2] = 'impossible'

      if possiblemove[0]>row and possiblemove[1]<column and possiblemove[2] != 'blank':
        for possible in possiblemoves:
          if possible[0]>possiblemove[0] and possible[1]<possiblemove[1]:
            possible[2] = 'impossible'
            
    possiblemoves = moves.removeimp(possiblemoves)
    return possiblemoves

  def knight(piece, board):
    row = piece[0]
    column = piece[1]
    possiblemoves = [[row+1, column+2], [row+1, column-2], [row-1, column-2], [row-1, column+2],[row+2, column-1], [row+2, column+1], [row-2, column+1], [row-2, column-1]] 
    for possiblemove in possiblemoves:
      possiblemove.append(moves.check(possiblemove, piece, board))
    possiblemoves = moves.removeimp(possiblemoves)
    return possiblemoves

  def king(piece, board):
    row = piece[0]
    column = piece[1]
    possiblemoves = [[row+1, column], [row+1, column+1], [row, column+1], [row-1, column+1],[row-1, column], [row-1, column-1], [row, column-1], [row+1, column-1]] 
    for possiblemove in possiblemoves:
      possiblemove.append(moves.check(possiblemove, piece, board))
    possiblemoves = moves.removeimp(possiblemoves)
    return possiblemoves      
  
  def queen(piece, board):
    possiblemoves = []
    possiblemoves.append(moves.rook(piece, board))
    possiblemoves.append(moves.bishop(piece, board))
    possiblemoves = moves.removeimp(possiblemoves)
    return possiblemoves

board = frameboard.createboard()



```
